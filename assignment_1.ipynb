## Theory: equations for potential flow:

the equations come from the assumption that the flow is **steady, incompressible, irrotational**, therefore 
- $ \nabla \cdot V  = 0 $      incompressible  
- $ \nabla \times V  = 0 $     irrotational 
- $ \frac{\partial V}{\partial t} = 0 $       steady

 
- Velocity in potential flow: $\mathbf{V} = \nabla \phi$, which can be rewritten in full form as 
$ V = \frac{\partial \phi}{\partial x}i + \frac{\partial \phi}{\partial y}j $

where $\phi$ is called POTENTIAL : similar to electromagnetism. 

NOTE: meaning of $\phi$ can be seen as elevation in a geographical map. it is a scalar function that has a gradient in both direction (imagine gradient search, drop a ball from mountain)

**only** in 2D, associated also to the stream function $\psi$

$$
\psi =

\begin{cases}
    u = \frac{\partial \psi}{\partial y} \\

    v = - \frac{\partial \psi}{\partial x}
\end{cases}

$$

important, we consider this as a property, but theoretically this equations are valid only for a set of potential which laplacian is null (in our case this will be conservation of mass)
$$\nabla^2 \phi = 0$$
